```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
def dat2array(file_path):
    data=[]
    with open(file_path, 'r', encoding='utf-8') as file: 
        
        for i,line in enumerate(file):
              a=line.replace(',','.').replace('\t', ';').replace('\n','')

              if i >= 39: data.append(a.split(';'))

    freqs = [float(val[0]) for val in data] 
    intensities = [float(val[1]) for val in data]

    return freqs,intensities
    
```

# SHG measurement of TIBI

```python
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(15, 5))  


colors = cm.plasma(np.linspace(0, 1, 13))
peak_list,tmp_list=[],[]


for idx,i in enumerate(np.arange(300, 440, 20)): 
    
    freqs,intensities = dat2array(f'TIBI_400uW_1400nm/{i}k')        #change path for other materials
    
    peak_list.append(np.max(intensities))
    tmp_list.append(i)

    #linker plot:
    ax0.plot(freqs, [val/44702 for val in intensities], color=colors[6+idx], label=f'{i}K')
    ax0.set_xlim(650,750)
    ax0.legend(loc='upper right')
    ax0.set_xlabel('Wavelength in nm')
    ax0.set_ylabel('SHG intensity in a.u.')

    #rechter plot:
    ax1.scatter(i,np.max(intensities)/44702, color=colors[6+idx])
    ax1.set_xlabel('Temperature in K')
    ax1.set_ylabel('Maximal SHG intensity in a.u.')


fig.suptitle('SHG measurement of TIBI at 1400nm and 400$\mu W ~~~ (300K \\rightarrow 420K )$', fontsize=18)

plt.savefig('./TIBI.png', dpi=600)

print(np.max(peak_list))
```
